## Pegasus

```python
peg_df=pd.read_csv('/content/drive/MyDrive/DISSERTATION/Dataset/test_peg_v4.csv')
```

```python
#Replacing the abstact without stopwords to the original abstract for evaluation
peg_df = peg_df.drop(columns=["abstract"])
peg_df["abstract"] = test_df["abstract"]
```

```python
df_copy=peg_df.copy()
```

```python
#Coherence Scoring Prompt
df_copy['Message']="Score the following summary given the corresponding article with respect to coherence from 0 to 1 with up to 4 decimal places. Coherence measures how well the summary logically follows from and clearly conveys the main ideas of the source article. 1 point indicates that the summary is perfectly coherent: it logically follows from the source article, clearly conveys the main ideas, and maintains a smooth and logical flow. 0 point indicates that the summary is completely incoherent: it does not logically follow from the source article and is disjointed or confusing.\nSummary: " + df_copy["generated_summary"] + "\nSource Article: " + df_copy["abstract"] + "\nExplain your reasoning step by step and then give a Score: "
```

```python
print(df_copy['Message'][0])
```

```python
#API Key
client = OpenAI(api_key="PASTE YOUR API KEY HERE")
```

```python
#GPT response generation
def get_response(message):
    stream = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": message}],
        stream=True,
    )
    response_content = ""
    for chunk in stream:
        if chunk.choices[0].delta.content is not None:
            response_content += chunk.choices[0].delta.content
    return response_content
```

```python
df_copy["Response"]=df_copy["Message"].apply(get_response)
```

```python
df_copy
```

```python
df_copy.to_csv('/content/drive/MyDrive/DISSERTATION/Results/Coherence/peg_v4_gpt2.csv', index=False)
```
