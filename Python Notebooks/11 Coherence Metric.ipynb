```python
import pandas as pd
import spacy
import numpy as np
import torch
import torch_geometric
from torch_geometric.data import Data,Batch
from torch_geometric.loader import DataLoader
from torch.nn import Linear, LSTM, Module
from torch_geometric.nn import RGCNConv, global_mean_pool,GCNConv
from allennlp.modules.elmo import Elmo, batch_to_ids
from sklearn.model_selection import train_test_split
from torch.optim.lr_scheduler import StepLR
import random
import torch.nn as nn
import torch.nn.functional as F
```

```python
# Load SpaCy model for entity extraction
nlp = spacy.load("en_core_web_sm")
```

```python
elmo_options_file = "https://allennlp.s3.amazonaws.com/models/elmo/2x4096_512_2048cnn_2xhighway/elmo_2x4096_512_2048cnn_2xhighway_options.json"
elmo_weights_file = "https://allennlp.s3.amazonaws.com/models/elmo/2x4096_512_2048cnn_2xhighway/elmo_2x4096_512_2048cnn_2xhighway_weights.hdf5"
elmo = Elmo(elmo_options_file, elmo_weights_file, 1, dropout=0)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
def create_entity_grid(sentences):
    doc = nlp(" ".join(sentences))
    entities = list(set(ent.text for ent in doc.ents))

    #Entity Grid
    grid = np.zeros((len(sentences), len(entities)), dtype=int)

    for i, sent in enumerate(sentences):
        for j, ent in enumerate(entities):
            if ent in sent:
                grid[i, j] = 1
    return grid, entities
```

```python
def grid_to_graph(grid):
    edge_index = []
    edge_type = []
    num_sentences, _ = grid.shape

    # Entity-based edges with type 1
    for i in range(num_sentences):
        for j in range(i + 1, num_sentences):
            if np.any(grid[i] & grid[j]):
                edge_index.append([i, j])
                edge_type.append(1)  # Indicating shared entities

    # Adjacency-based edges with type 0
    for i in range(num_sentences - 1):
        edge_index.append([i, i + 1])
        edge_type.append(0)  # Indicating adjacency

    edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()
    edge_type = torch.tensor(edge_type, dtype=torch.long)
    return edge_index, edge_type
```

```python
#Self Attention Layer
class SelfAttentionLayer(nn.Module):
    def __init__(self, input_dim):
        super(SelfAttentionLayer, self).__init__()
        self.query = nn.Linear(input_dim, input_dim)
        self.key = nn.Linear(input_dim, input_dim)
        self.value = nn.Linear(input_dim, input_dim)
        self.scale = np.sqrt(input_dim)

    def forward(self, x):
        q = self.query(x)
        k = self.key(x)
        v = self.value(x)

        scores = torch.matmul(q, k.transpose(-2, -1)) / self.scale
        attention_weights = F.softmax(scores, dim=-1)
        attended = torch.matmul(attention_weights, v)
        return attended
```

```python
#BiLSTM Layer
class BiLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers=1, dropout=0.0):
        super(BiLSTM, self).__init__()
        self.lstm = nn.LSTM(
            input_dim, hidden_dim, num_layers=num_layers,
            bidirectional=True, dropout=dropout, batch_first=True
        )

    def forward(self, x):
        self.lstm.flatten_parameters()
        out, _ = self.lstm(x)
        return out
```

```python
#RGCN model initialization
class GCNWithAttention(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_relations):
        super(GCNWithAttention, self).__init__()
        self.conv1 = RGCNConv(input_dim, hidden_dim, num_relations)
        self.conv2 = RGCNConv(hidden_dim, hidden_dim, num_relations)
        self.bn1 = nn.BatchNorm1d(hidden_dim)
        self.relu = nn.LeakyReLU()
        self.attention = SelfAttentionLayer(hidden_dim)
        self.fc = nn.Linear(hidden_dim, 1)  # Final layer for binary classification

    def forward(self, x, edge_index, edge_type, batch):
        x = self.conv1(x, edge_index, edge_type)
        x = self.bn1(x)
        x = self.relu(x)
        x = self.conv2(x, edge_index, edge_type)
        x = self.attention(x)  # Apply self-attention here
        x = global_mean_pool(x, batch)
        x = self.fc(x)
        return x
```

```python
def compute_elmo_embeddings(sentences):
    character_ids = batch_to_ids([sent.split() for sent in sentences])
    embeddings = elmo(character_ids)
    return embeddings['elmo_representations'][0].detach()
```

```python
def create_graph_data(sentences, label, batch_idx, bilstm, device, return_sentences=False):
    grid, entities = create_entity_grid(sentences)
    edge_index, edge_type = grid_to_graph(grid)

    elmo_embeddings = compute_elmo_embeddings(sentences)
    bilstm_input = elmo_embeddings.to(device)
    lstm_out = bilstm(bilstm_input).detach()

    x = lstm_out.mean(dim=1).detach()

    y = torch.tensor([label], dtype=torch.float, device=device)
    batch = torch.full((x.shape[0],), batch_idx, dtype=torch.long, device=device)

    data = Data(x=x, edge_index=edge_index, edge_type=edge_type, y=y, batch=batch)

    if return_sentences:
        return data, sentences
    else:
        return data
```

```python
elmo_embedding_dim = 1024
bilstm_hidden_dim = 64
gcn_hidden_dim = 64

#BiLSTM and GCN with Attention
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
bilstm = BiLSTM(input_dim=elmo_embedding_dim, hidden_dim=bilstm_hidden_dim).to(device)
gcn_with_attention = GCNWithAttention(input_dim=2 * bilstm_hidden_dim, hidden_dim=gcn_hidden_dim, num_relations=2).to(device)

# Optimizer and Loss Function
optimizer = torch.optim.Adam(list(bilstm.parameters()) + list(gcn_with_attention.parameters()), lr=0.01)
loss_fn = torch.nn.BCEWithLogitsLoss()
```

## Pegasus

```python
peg_df = pd.read_csv('/content/drive/MyDrive/DISSERTATION/Dataset/test_peg_v4.csv')
```

```python
train_summaries, test_summaries = train_test_split(peg_df['generated_summary'], test_size=0.2, random_state=42)
```

```python
#Create graph from train data
train_data = []
for i, summary in enumerate(train_summaries):
    doc = nlp(summary)
    sentences = [sent.text for sent in doc.sents]
    train_data.append(create_graph_data(sentences, label=1, batch_idx=i * 2, bilstm=bilstm, device=device))
    shuffled_sentences = random.sample(sentences, len(sentences))
    train_data.append(create_graph_data(shuffled_sentences, label=0, batch_idx=i * 2 + 1, bilstm=bilstm, device=device))
```

```python
test_data = []
test_sentences = []
for i, summary in enumerate(test_summaries):
    doc = nlp(summary)
    sentences = [sent.text for sent in doc.sents]
    data, original_sentences = create_graph_data(sentences, label=1, batch_idx=i, bilstm=bilstm, device=device, return_sentences=True)
    test_data.append(data)
    test_sentences.append(original_sentences)
```

```python
# Function to train the model
def train_model(train_loader, model, optimizer, loss_fn, device, epochs=10):
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for data in train_loader:
            data = data.to(device)
            optimizer.zero_grad()
            out = model(data.x, data.edge_index, data.edge_type, data.batch).view(-1)
            loss = loss_fn(out, data.y)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch+1}, Loss: {total_loss / len(train_loader)}")
```

```python
# Function to evaluate the model
def evaluate_model(test_loader, model):
    model.eval()
    scores = []
    true_labels = []
    with torch.no_grad():
        for data in test_loader:
            data = data.to(device)
            out = model(data.x, data.edge_index, data.edge_type, data.batch).view(-1)
            scores.extend(torch.sigmoid(out).tolist())
            true_labels.extend(data.y.tolist())
    return scores, true_labels
```

```python
train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)
```

```python
train_model(train_loader, gcn_with_attention, optimizer, loss_fn, device)
```

```python
#Predict the coherence scores
coherence_scores, true_labels = evaluate_model(test_loader, gcn_with_attention)
```

```python
peg_avg_coherence_score = np.mean(coherence_scores)
print(f'Average Coherence Score: {peg_avg_coherence_score}')
```

```python
results_df = pd.DataFrame({
    'summary': test_sentences,
    'coherence_score': coherence_scores,
})
```

```python
results_df['summary'] = results_df['summary'].apply(lambda sents: ' '.join(sents))
```

```python
print(results_df)
```

```python
results_df.to_csv('/content/drive/MyDrive/DISSERTATION/Results/Coherence/peg_v4_new.csv', index=False)
```
