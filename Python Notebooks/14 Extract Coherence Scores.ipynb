```python
import pandas as pd
import re
```

```python
#Extracts the last decimal place
def extract_score_decimal(text):
    match = re.search(r'(?i)score[:\s]*\d+\.\d+', text)
    if match:
        return float(re.findall(r'\d+\.\d+', match.group())[0])
    return None
```

## Pegasus

```python
peg_df=pd.read_csv('/content/drive/MyDrive/DISSERTATION/Results/Coherence/peg_v4_gpt2.csv')
```

```python
peg_df['Answer'] = peg_df['Response'].apply(extract_score_decimal)
```

```python
peg_df.to_csv('/content/drive/MyDrive/DISSERTATION/Results/Coherence/Extracted/peg_v4_gpt.csv', index=False)
```

```python
print(f'Average Coherence score for Pegasus: {peg_df["Answer"].mean()}')
```
