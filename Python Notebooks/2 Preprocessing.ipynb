```python
import pandas as pd
import spacy
```

```python
#Spacy library for stopwords
nlp = spacy.load('en_core_web_sm')
```

## TRAIN DATASET

```python
train_df=pd.read_csv('/content/drive/MyDrive/DISSERTATION/Dataset/Extracted Dataset (csv)/train_plos.csv')
```

```python
train_df
```

### Abstract

```python
#Replace the last newline character in abstract
train_df['abstract']=train_df['abstract'].str.replace('\n','')
```

```python
train_df.iloc[0]['abstract']
```

```python
#Remove stopwords from the abstract
def rem_stopW(text):
  doc=nlp(text)
  new_text=' '.join([token.text for token in doc if not token.is_stop])
  return new_text

train_df['abstract']=train_df['abstract'].apply(rem_stopW)
```

### Plain Language Summary (PLS)

```python
#Replace the last newline character in PLS
train_df['plain language summary']=train_df['plain language summary'].str.replace('\n','')
```

```python
train_df.iloc[1]['plain language summary']
```

```python
#Converting to CSV
train_df.to_csv('/content/drive/MyDrive/DISSERTATION/Dataset/Preprocessed Dataset/train_plos.csv')
```
